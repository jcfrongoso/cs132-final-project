```python
import pandas as pd
import numpy as np
import os
import ipywidgets as widgets
from matplotlib import pyplot as plt
from ipywidgets import interact
```

```python
farmgate = './Dataset/Fruits Farmgate Prices by Commodity, by Region and by Province, Monthly, 2010-2024.csv'
retail = './Dataset/Fruits Retail Prices of Agricultural Commodities by Geolocation, Commodity, Year and Period.csv'

farmgate = pd.read_csv(farmgate, skiprows=2)
retail = pd.read_csv(retail, skiprows=2)
```

```python
def changeToNull(x):
    if x == '..':
        return np.nan
    return x

for i in farmgate.columns:
    farmgate[i] = farmgate[i].apply(changeToNull)
```

```python
farmgate.info()
```

```python
farmgate.dropna(axis=0, how='any', inplace=True)
```
