```python
import pandas as pd
import numpy as np
import os
```

```python
dir = './Dataset/Approved Investments.csv'

ai_df = pd.read_csv(dir, skiprows=2)
ai_df.head()
```

```python
ai_df = ai_df.melt('Country')
ai_df.columns = ['Country', 'Quarter', 'Investment']
```

```python
ai_df.head()
```

```python
ai_df.info()
```

```python
ai_df.describe()
```

```python
def get_second_value(val):
    return val.split()[1]

ai_df['Quarter'] = ai_df['Quarter'].apply(get_second_value)
```

```python
from matplotlib import pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def plot(country):
    filtered = ai_df[ai_df['Country'] == country]
    plt.plot(filtered['Quarter'], filtered['Investment'])
    plt.show()

countries = list(ai_df['Country'].unique())

x = widgets.Dropdown(options=countries)

interact(plot, country=x)
```
