```python
import pandas as pd
import os
import numpy as np
```

```python
dir = './Dataset/'
volume_ds = 'Aquaculture Volume of Production by Species, Geolocation, Year and Quarter.csv'
value_ds = 'Aquaculture Value of Production by Species, Geolocation, Year and Quarter.csv'
```

```python
volume_df = pd.read_csv(f'{dir}{volume_ds}', skiprows=2)
volume_df.head()
```

```python
volume_df = volume_df.melt(['Species', 'Geolocation'], var_name="Year-Quarter", value_name="Volume", ignore_index=True)
```

```python
value_df = pd.read_csv(f'{dir}{value_ds}', skiprows=2)
value_df.head()
```

```python
value_df = value_df.melt(['Species', 'Geolocation'], var_name="Year-Quarter", value_name="Value", ignore_index=True)
```

```python
merged_df = volume_df.copy()
merged_df = pd.merge(merged_df, value_df, on=['Species', 'Geolocation', 'Year-Quarter'], how='inner')
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df.to_csv('Aquaculture Value and Volume of Production - Merged CSV.csv')
```
