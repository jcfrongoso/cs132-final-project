```python
import pandas as pd
import os
import numpy as np
```

```python
dir = './Dataset/'
dir_upd = os.listdir(dir)
```

```python
for file in dir_upd:
    df = pd.read_csv(f'{dir}{file}', skiprows=1)
    print(df.head())
```

```python
cpi_df = pd.read_csv(f'{dir}{dir_upd[0]}', skiprows=1)
purchasing_power_df = pd.read_csv(f'{dir}{dir_upd[1]}', skiprows=1)
```

## CPI

```python
cpi_df.head()
```

```python
cpi_df.pivot(index=['Geolocation'], columns=['Commodity Description'])
```

```python
dates = (cpi_df.columns.to_list())[2:]
cpi_df.melt(id_vars=['Geolocation', 'Commodity Description'], value_vars=dates)
```

## Purchasing Power

```python
purchasing_power_df.head()
```

```python
dates = (purchasing_power_df.columns.to_list())[2:]
purchasing_power_df.melt(id_vars=['Geolocation'], value_vars=dates)
```

```python
from matplotlib import pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

pivot_ppdf = purchasing_power_df.melt(id_vars=['Geolocation'], value_vars=dates).copy()

geolocations = pivot_ppdf['Geolocation'].unique()

v = widgets.Dropdown(
    options=geolocations,
    description="Locations"
)

def func(loc):
    filtered_by_loc = pivot_ppdf[pivot_ppdf['Geolocation'] == loc]
    plt.plot(filtered_by_loc['variable'], filtered_by_loc['value'])
    plt.show()

interact(func, loc=v)
```
