```python
import pandas as pd
import os
import numpy as np
```

```python
dir = './Dataset/'
dir_upd = os.listdir(dir)
```

```python
for file in dir_upd:
    df = pd.read_csv(f'{dir}{file}', skiprows=1)
    print(df.head())
```

```python
cpi_df = pd.read_csv(f'{dir}{dir_upd[1]}', skiprows=1)
purchasing_power_df = pd.read_csv(f'{dir}{dir_upd[2]}', skiprows=1)
```

## CPI

```python
cpi_df.head()
```

```python
cpi_df.pivot(index=['Geolocation'], columns=['Commodity Description'])
```

```python
dates = (cpi_df.columns.to_list())[2:]
cpi_df.melt(id_vars=['Geolocation', 'Commodity Description'], value_vars=dates)
```

```python
reformatted_cpi_df = cpi_df.melt(id_vars=['Geolocation', 'Commodity Description'], value_vars=dates).copy()
```

```python
reformatted_cpi_df.head()
```

```python
def remove_dots(string):
    if string.find('.') == -1:
        return string
    return string[len(string)-string[::-1].find('.'):]

# print(remove_dots('.wow'))
reformatted_cpi_df['Geolocation'] = reformatted_cpi_df['Geolocation'].apply(remove_dots)
```

```python
from matplotlib import pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

geolocation = list(reformatted_cpi_df['Geolocation'].unique())
commodity = list(reformatted_cpi_df['Commodity Description'].unique())

def plot_time_series(geol, comm):
    filtered_df = reformatted_cpi_df[(reformatted_cpi_df['Geolocation'] == geol) & (reformatted_cpi_df['Commodity Description'] == comm)]
    plt.plot(filtered_df['variable'], filtered_df['value'])

x = widgets.Dropdown(options=geolocation)
y = widgets.Dropdown(options=commodity)
interact(plot_time_series, geol=x, comm=y)
plt.show()
```

## Purchasing Power

```python
purchasing_power_df.head()
```

```python
dates = (purchasing_power_df.columns.to_list())[2:]
purchasing_power_df.melt(id_vars=['Geolocation'], value_vars=dates)
```

```python
from matplotlib import pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

pivot_ppdf = purchasing_power_df.melt(id_vars=['Geolocation'], value_vars=dates).copy()

geolocations = pivot_ppdf['Geolocation'].unique()

v = widgets.Dropdown(
    options=geolocations,
    description="Locations"
)

def func(loc):
    filtered_by_loc = pivot_ppdf[pivot_ppdf['Geolocation'] == loc]
    plt.plot(filtered_by_loc['variable'], filtered_by_loc['value'])
    plt.show()

interact(func, loc=v)
```

```python
reg_three = pd.read_csv('./Dataset/Consumer Price Index for All Income Households by Commodity Group Region III Food Only.csv', skiprows=2, encoding='latin-1')
reg_three.head()
```

```python
reg_three.shape
```

```python
reg_three_upd = reg_three.melt(['Geolocation', 'Commodity Description'], var_name='Month/Year', value_name='CPI')
reg_three_upd.head()
```

```python
def plot_graph(geol, comm):
    filtered = reg_three_upd[(reg_three_upd['Geolocation'] == geol) & (reg_three_upd['Commodity Description'] == comm)]
    plt.plot(filtered['Month/Year'], filtered['CPI'])
    plt.xticks(rotation=45)
    plt.show()
geolocation = list(reg_three_upd['Geolocation'].unique())
commodity = list(reg_three_upd['Commodity Description'].unique())
x = widgets.Dropdown(options=geolocation)
y = widgets.Dropdown(options=commodity)
interact(plot_graph, geol=x, comm=y)
```
